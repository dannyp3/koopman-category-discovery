### Helper Functions

```python
def get_lambda_source(lambda_func):
    try:
        source_lines = inspect.getsource(lambda_func)
        return source_lines.strip()
    except OSError:
        return None
```

```python
def make_functions(rng):
    f1  = (lambda s=rng.uniform(0.8, 1.2): lambda x: x / 5 * s)()
    f2  = (lambda s=rng.uniform(0.8, 1.2): lambda x: x**0.4 * s)()
    f3  = (lambda a=rng.uniform(0, 1):     lambda x: np.sqrt(x + a))()
    f4  = (lambda a=rng.uniform(0, 2):     lambda x: np.log(x + 1 + a))()
    f5  = (lambda k=rng.uniform(0.005, 0.02): lambda x: np.exp(-k * x))()
    f6  = (lambda m=rng.uniform(40, 60), s=rng.uniform(8, 15):
           lambda x: np.exp(-((x - m) / s)**2))()
    f7  = (lambda k=rng.uniform(0.05, 0.15): lambda x: np.sin(k * x))()
    f8  = (lambda k=rng.uniform(0.05, 0.15): lambda x: np.cos(k * x))()
    f9  = (lambda k=rng.uniform(0.08, 0.12), d=rng.uniform(0.005, 0.02):
           lambda x: np.sin(k * x) * np.exp(-d * x))()
    f10 = (lambda d=rng.uniform(0.03, 0.07): lambda x: x * np.exp(-d * x))()

    f11 = (lambda a=rng.uniform(0, 1), k=rng.uniform(0.15, 0.25):
           lambda x: np.log(x + 1 + a) * np.sin(k * x))()
    f12 = (lambda m=rng.uniform(45, 55), s=rng.uniform(8, 15):
           lambda x: np.tanh((x - m) / s))()
    f13 = (lambda d=rng.uniform(4, 6):      lambda x: np.sin(x / d)**2)()
    f14 = (lambda a=rng.uniform(0, 1):      lambda x: 1 / (1 + x + a))()
    f15 = (lambda a=rng.uniform(0, 2):      lambda x: np.exp(-np.sqrt(x + a)))()
    f16 = (lambda d=rng.uniform(2, 5), k=rng.uniform(0.005, 0.02):
           lambda x: np.sin(x / d) + k * x)()
    f17 = (lambda d=rng.uniform(3, 5), k=rng.uniform(0.005, 0.02):
           lambda x: np.cos(x / d) * np.exp(-k * x))()
    f18 = (lambda k1=rng.uniform(0.15, 0.25), k2=rng.uniform(0.03, 0.07):
           lambda x: np.sin(k1 * x) * np.sin(k2 * x))()
    f19 = (lambda m1=rng.uniform(25, 35), m2=rng.uniform(65, 75),
                 s=rng.uniform(4, 6):
           lambda x: np.exp(-((x - m1) / s)**2) + np.exp(-((x - m2) / s)**2))()
    f20 = (lambda k=rng.uniform(0.008, 0.015): lambda x: np.sinh(k * x))()

    funcs = [f1, f2, f3, f4, f5, f6, f7, f8, f9, f10,
             f11, f12, f13, f14, f15, f16, f17, f18, f19, f20]

    return {f"f{i+1}": fn for i, fn in enumerate(funcs)}

random_functions = make_functions(rng)
```

```python
def create_time_series(domain,random_functions,fs,plot=False):

    x_values = []
    for f in fs:
        x_values.append(random_functions[f](domain))
    
    X = np.vstack(x_values).T

    return X
```

```python
def dmd_by_hand(X_full,rank):
    m, n = X_full.shape

    if rank < 0 or rank > m:
        raise ValueError(f'Given rank is {rank} yet data given has dimension {m}')
    
    X = X_full[:,:-1]
    Y = X_full[:,1:]
    
    # SVD of X
    U, s, Vh = np.linalg.svd(X, full_matrices=False)
    
    # r rank of svd of X
    U_r = U[:,:rank]
    s_r = s[:rank]
    Vh_r = Vh[:rank,:]
    
    # Calculate A_tilde
    U_r_star = U_r.conj().T
    V_r = Vh_r.conj().T
    S_r_inv = np.diag(1.0 / s_r)
    
    A_tilde = U_r_star @ Y @ V_r @ S_r_inv
    
    # Eigendecomposition of A_tilde
    eigs, W = np.linalg.eig(A_tilde)
    
    # Koopman Modes & amplitudes
    modes = Y @ V_r @ S_r_inv @ W
    b, _, _, _ = np.linalg.lstsq(modes,X[:,0])

    return eigs, modes, b
```

### Upload Dataset

```python
num_cats = 5
num_samples = 100
path = rf"data\1-dimensional-systems\dataset_{num_cats}_class_{num_samples}_samples.pkl"
with open(path, 'rb') as f:
    dataset = pickle.load(f)


cats = list(dataset.keys())

m, n = dataset[cats[0]][0]['y'].shape
delay_embeddings = 3
num_observables = m * (delay_embeddings + 1)
num_segments = 5

if np.mod(n,num_segments) != 0:
    raise ValueError(f'Number of segments {num_segments} must divide given data size {n-delay_embeddings}')

segment_length = int(n/num_segments)
```

```python
# Turn this into a function and create a train and test split of different sizes

all_data = []
all_eigs = []
all_modes = []
all_amps = []

total_dmd_calculations = num_cats * num_samples * num_segments
print(f'Generating {total_dmd_calculations} DMD eigs/modes each with dimensionality {num_observables}')


for cat in cats:

    curr_data = dataset[cat]

    for index in range(num_samples):

        X = curr_data[index]['y'].T
        
        if delay_embeddings > 0:
            X = np.hstack([X[i:n-delay_embeddings+i,:] for i in range(delay_embeddings+1)])
        
        # DMD
        for sp in range(num_segments):
            start_ind = sp * segment_length
            end_ind = (sp + 1) * segment_length
            X_split = X[start_ind:end_ind,:]
            
            eigs, modes, b = dmd_by_hand(X_split.T, rank=num_observables)
            all_eigs.append(eigs)
            all_modes.append(modes)
            all_amps.append(b)

            all_data.append(X_split)

if len(all_eigs) != total_dmd_calculations:
    raise ValueError(f'Incorrect Number of DMD Run: expected {total_dmd_calculations}, but ran {len(all_eigs)}')

if all_eigs[0].shape[0] != num_observables:
    raise ValueError(f'Incorrect Observables Count: expected {num_observables}, but got {all_eigs[0].shape[0]}')
```

### Clustering

```python
def compute_wasserstein_metric(l1,l2,m1,m2,num_observables,q=2):
    
    # --- Step 1: Sample Inputs ---
    n = num_observables
    n_bar = num_observables

    m1 = m1 / np.sum(m1)
    m2 = m2 / np.sum(m2)
    
    assert np.isclose(np.sum(m1), 1.0)
    assert np.isclose(np.sum(m2), 1.0)
    
    # --- Step 2: Compute cost matrix C (shape n x n_bar) ---
    C = np.linalg.norm(l1[:, None, :] - l2[None, :, :], axis=2) ** 2
    # Flatten to 1D for linprog
    c = C.flatten()  # size (n * n_bar,)
    
    # --- Step 3: Equality Constraints ---
    
    # Total number of variables
    N = n * n_bar
    
    # 1. Row sum constraints: each row i must sum to m[i]
    A_eq_rows = np.zeros((n, N))
    for i in range(n):
        for j in range(n_bar):
            A_eq_rows[i, i * n_bar + j] = 1
    b_eq_rows = m1
    
    # 2. Column sum constraints: each column j must sum to m_bar[j]
    A_eq_cols = np.zeros((n_bar, N))
    for j in range(n_bar):
        for i in range(n):
            A_eq_cols[j, i * n_bar + j] = 1
    b_eq_cols = m2
    
    # 3. Total mass constraint
    A_eq_total = np.ones((1, N))
    b_eq_total = np.array([1.0])
    
    # Combine constraints
    A_eq = np.vstack([A_eq_rows, A_eq_cols, A_eq_total])
    b_eq = np.concatenate([b_eq_rows, b_eq_cols, b_eq_total])
    
    # --- Step 4: Bounds ---
    bounds = [(0, None)] * N  # rho_ij >= 0
    
    # --- Step 5: Solve ---
    result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')
    
    # --- Step 6: Extract solution ---
    if result.success:
        rho_star = result.x.reshape((n, n_bar))
        transport_cost = np.sum(rho_star * C)
        # print("Optimal transport plan *:\n", rho_star)
        # print("Transport cost (before q-root):", transport_cost)

        wasserstein_metric = transport_cost ** (1 / q)
        # print(f"Wasserstein-like metric (1/{q} root):", wasserstein_metric)
    else:
        print("Optimization failed:", result.message)

    return wasserstein_metric
```

```python
downsampled_calculations = df_sample.shape[0]
metric_matrix = np.zeros((downsampled_calculations,downsampled_calculations))

total_metric_calculations = int(downsampled_calculations**2/2)

print(f'Generating (1/2) * {downsampled_calculations}^2 = {total_metric_calculations} Wasserstein distance metrics')

# Columns for extracting data
eig_columns = [col for col in df.columns if 'eig' in col]
norm_mode_columns = [col for col in df.columns if 'norm_mode' in col]

q = 2
running_sum = 0
for i in range(downsampled_calculations):
    for j in range(i):

        # Eigenvalues
        l1 = df_sample.loc[i,eig_columns].values[:,np.newaxis]
        l2 = df_sample.loc[j,eig_columns].values[:,np.newaxis]
        
        # Mass vectors (normed modes)
        m1 = df_sample.loc[i,norm_mode_columns].values.real
        m2 = df_sample.loc[j,norm_mode_columns].values.real

        # Compute Wasserstein Metric
        metric_matrix[i,j] = compute_wasserstein_metric(l1,l2,m1,m2,num_observables,q)
        
        running_sum += 1

        if np.mod(running_sum,int(total_metric_calculations/20)) == 0:
            print(f'{int(running_sum/int(total_metric_calculations/100))}%')
```

```python
full_metric_matrix = metric_matrix + metric_matrix.T
plt.imshow(full_metric_matrix)
plt.colorbar()
plt.title('Wasserstein Divergence Distance Matrix')
plt.show()
```

```python
# Test which embedding space dimension minimizes metric matrix reconstruction error
max_dim = 10
num_tests = 3

errors = np.zeros((num_tests,max_dim))

for i in range(max_dim):
    for j in tqdm(range(num_tests),desc=f'Percent Error for {i+1}-d Euclidean Representation'):
        # Create Euclidean Representation Based off Distance Matrix
        embedding = MDS(n_components=i+1, dissimilarity='precomputed')
        X_transformed = embedding.fit_transform(full_metric_matrix)
        
        # Reconstruct Distance Matrix
        D_reconstructed = squareform(pdist(X_transformed))
    
        errors[j,i] = np.linalg.norm((D_reconstructed - full_metric_matrix))
        # percent_errors[j,i] = np.nanmean(abs(D_reconstructed - full_metric_matrix) / full_metric_matrix)
```

```python
plt.figure(figsize=(6,3.5))
plt.plot(np.arange(1,max_dim+1),errors.mean(axis=0),label='Mean')
plt.fill_between(np.arange(1,max_dim+1),errors.min(axis=0),errors.max(axis=0),alpha=0.3,color='red',label='Full Range')
plt.xlabel('Euclidean Dimension')
plt.ylabel('Percent Error')
plt.title('Distance Matrix Reconstruction Error')
# plt.ylim([0,100])
plt.xticks(np.arange(1,max_dim+1,dtype=int))
# plt.yticks(np.arange(0,110,10,dtype=int))
plt.legend()
plt.grid()
plt.show()
```

### Generate Codebook

```python
k = 5
centroid, label, inertia = k_means(X_transformed, k, random_state=SEED)

# Find the 5 closest points in the training data to the cluster centers (surrogate cluster centers)
distances = cdist(centroid,X_transformed)
cluster_indices = np.argmin(distances, axis=1)
cluster_centers = X_transformed[cluster_indices,:]

# Use those surrogate cluster centers to make the codebook dataframe
codebook = df_sample.loc[cluster_indices].reset_index(drop=True)

# Reassign the labels for the downsampled dataframe based on surrogate cluster centers
distances = cdist(cluster_centers,X_transformed)
sample_label = np.argmin(distances,axis=0)

# Compare percent change in labels from using surrogate cluster centers
tol = 1e-8
num_changed = np.sum(abs(sample_label - label) >= tol)
print(f'{num_changed}/{df_sample.shape[0]} labels changed ({round(num_changed/df_sample.shape[0] * 100,2)}%)')
```

```python
train_size = df_train.shape[0]
training_metric_matrix = np.zeros((k,train_size))

total_training_metric_calculations = int(k * train_size)

print(f'Generating {k} * {train_size} = {total_training_metric_calculations} Wasserstein distance metrics')

# Columns for extracting data
eig_columns = [col for col in df.columns if 'eig' in col]
norm_mode_columns = [col for col in df.columns if 'norm_mode' in col]

q = 2
running_sum = 0
for i in codebook.index:
    for j in df_train.index:

        # Eigenvalues
        l1 = codebook.loc[i,eig_columns].values[:,np.newaxis]
        l2 = df_train.loc[j,eig_columns].values[:,np.newaxis]
        
        # Mass vectors (normed modes)
        m1 = codebook.loc[i,norm_mode_columns].values.real
        m2 = df_train.loc[j,norm_mode_columns].values.real

        # Compute Wasserstein Metric
        training_metric_matrix[i,j] = compute_wasserstein_metric(l1,l2,m1,m2,num_observables,q)
        
        running_sum += 1

        if np.mod(running_sum,int(total_training_metric_calculations/20)) == 0:
            print(f'{int(running_sum/int(total_training_metric_calculations/100))}%')
```

### Classification

```python
test_size = df_test.shape[0]
test_metric_matrix = np.zeros((k,test_size))

total_test_metric_calculations = int(k * test_size)

# Columns for extracting data
eig_columns = [col for col in df.columns if 'eig' in col]
norm_mode_columns = [col for col in df.columns if 'norm_mode' in col]

q = 2
running_sum = 0
for i in codebook.index:
    for j in df_test.index:

        # Eigenvalues
        l1 = codebook.loc[i,eig_columns].values[:,np.newaxis]
        l2 = df_test.loc[j,eig_columns].values[:,np.newaxis]
        
        # Mass vectors (normed modes)
        m1 = codebook.loc[i,norm_mode_columns].values.real
        m2 = df_test.loc[j,norm_mode_columns].values.real

        # Compute Wasserstein Metric
        test_metric_matrix[i,j] = compute_wasserstein_metric(l1,l2,m1,m2,num_observables,q)
        
        running_sum += 1

        if np.mod(running_sum,int(total_test_metric_calculations/20)) == 0:
            print(f'{int(running_sum/int(total_test_metric_calculations/100))}%')
```

```python
# Identify number of samples from each system category in both training and testing samples
num_train_samples = int(len(df_train.groupby(['target','sample']).groups) / num_cats)
num_test_samples = int(len(df_test.groupby(['target','sample']).groups) / num_cats)

# Create cluster assignments for each segment in training and testing set
training_cluster_assignments = np.argmin(training_metric_matrix,axis=0).reshape(num_cats,num_train_samples,num_segments)
test_cluster_assignments = np.argmin(test_metric_matrix,axis=0).reshape(num_cats,num_test_samples,num_segments)

# Count number of each cluster assignment in training set
N_ks = np.array([(np.sum(training_cluster_assignments == i,axis=2) > 0).ravel().sum() for i in range(k)])
N = num_cats * num_samples # Total number of time series samples

# Create histogram of cluster assignments for training data
n_train_matrix = np.concatenate([np.sum(training_cluster_assignments == i,axis=2).ravel() for i in range(k)]).reshape(k,num_cats*num_train_samples).T
c_train_matrix = n_train_matrix / num_segments
inv_c_train_matrix = c_train_matrix * np.log(N / N_ks)

# Create histogram of cluster assignments for testing data
n_test_matrix = np.concatenate([np.sum(test_cluster_assignments == i,axis=2).ravel() for i in range(k)]).reshape(k,num_cats*num_test_samples).T
c_test_matrix = n_test_matrix / num_segments
inv_c_test_matrix = c_test_matrix * np.log(N / N_ks)

# Extract target assignments (classification labels)
train_target = df_train['target'].values[-num_segments::-num_segments][-1::-1]
test_target = df_test['target'].values[-num_segments::-num_segments][-1::-1]
```

```python
max_x = 0
max_v = 0
for i in range(len(all_data)):
    x = max(abs(all_data[i][:,0]))
    v = max(abs(all_data[i][:,1]))
    if x > max_x:
        max_x = x
    if v > max_v:
        max_v = v
max_x, max_v
```

```python
# Plot Phase Diagrams for Full Test Set
test_index = 0

for s, system in enumerate(systems):
    
    print(f'Creating Phase Diagrams for {system}: ', end='')
    
    figs, axes = plt.subplots(5,4,figsize=(14,14))
    ax = axes.ravel()
    
    for ind in range(20):
        
        # Identify target & sample number
        target = test_sets.loc[test_index,'target']
        sample = test_sets.loc[test_index,'sample']
        
        # Identify segment indices in all_data
        start = 500 * target + 5 * sample
        segment_indices = np.arange(start,start+5)
        
        start = 100
        for i,d in enumerate(segment_indices):
        
            x, dx = all_data[d][:,:2].T
            
            ax[ind].axhline(y=0, color='k', linewidth=2)
            ax[ind].axvline(x=0, color='k', linewidth=2)
            ax[ind].plot(x,dx,label=f'Segment {i+1}')
            for j in range(0, len(x)-1, 50):
                ax[ind].annotate("",
                             xy=(x[j+1], dx[j+1]),
                             xytext=(x[j], dx[j]),
                             arrowprops=dict(arrowstyle="->", color="blue", lw=1))
        
        title = [f'True: {systems[target]}',
                 f'Pred: {systems[y_pred[test_index]]}']
    
        if test_target[test_index] == y_pred[test_index]:
            ax[ind].set_facecolor('palegreen')
        else:
            ax[ind].set_facecolor('lightcoral')
        
        ax[ind].set_title('\n'.join(title))
        ax[ind].set_xlim([-np.ceil(abs(max_x)),np.ceil(abs(max_x))])
        ax[ind].set_ylim([-np.ceil(max_v),np.ceil(max_v)])
        ax[ind].grid()
        ax[ind].legend(fontsize=6)

        test_index += 1

    nrows, ncols = 5, 4

    for i, a in enumerate(ax):
        row = i // ncols
        col = i % ncols
    
        if row == nrows - 1:  # bottom row
            a.set_xlabel("x")
    
        if col == 0:  # leftmost column
            a.set_ylabel("dx/dt")

    
    plt.suptitle(f'{system} Phase Diagrams\n',fontsize=20)
    plt.tight_layout()
    plt.savefig(f'plots\\{num_cats}_systems\\results_{cats[s]}.png', dpi=300)
    
    print('Done!')
```
